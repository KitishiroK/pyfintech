## GoodInfo 表格抓取

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://goodinfo.tw/StockInfo/StockBzPerformance.asp?STOCK_ID=2330')
res.encoding = 'utf-8'
soup = BeautifulSoup(res.text )
# 表格資料放置於 id=divFinDetail 之中
#print soup.select('#divFinDetail table')
```

```python
import pandas
#使用pandas 的 read_html 將表格資料讀取至DataFrame Lists
dfs = pandas.read_html(soup.select('#divFinDetail table')[0].prettify('utf-8'), encoding = 'utf-8')
# 取得Lists 理面的第一個元素
dfs[0]
```

## 各產業EPS統計資訊

```python
import requests
from bs4 import BeautifulSoup
payload = {
'encodeURIComponent':'1',
'step':'1',
'firstin':'1',
'TYPEK':'sii',
'code':'24',
'year':'104',
'season':'01'    
}
res = requests.post('http://mops.twse.com.tw/mops/web/ajax_t163sb19', data=payload)
res.encoding = 'utf-8'

soup = BeautifulSoup(res.text)
```

```python
import pandas
#使用pandas 的 read_html 將表格資料讀取至DataFrame Lists
dfs = pandas.read_html(soup.select('.hasBorder')[0].prettify('utf-8'), encoding = 'utf-8', header=0)
# 取得Lists 理面的第一個元素
dfs[0]
```
